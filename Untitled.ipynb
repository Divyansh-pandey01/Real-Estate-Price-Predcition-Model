```python
def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    
    return data.iloc[train_indices], data.iloc[test_indices]

```

```python

train_set, test_set = split_train_test(housing,0.2)
```

## Creating a Pipeline



```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    #    ..... add as many as you want in your pipeline
    ('std_scaler',StandardScaler()),
])


```

```python
housing = housing.copy()
```

```python

# Drop columns with all NaN values
housing = housing.dropna(axis=1, how='all')
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr.shape
```

```python

##Selecting a desired model for AS Real Estates
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model= DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
#Testing the model on test data

import numpy as np
from sklearn.metrics import mean_squared_error

# Step 1
x_test = strat_test_set.drop("MEDV", axis=1, errors='ignore')
y_test = strat_test_set["MEDV"].copy()

# Step 2: remove unwanted columns
x_test = x_test.loc[:, ~x_test.columns.str.contains('^Unnamed')]

# Step 3: match columns
train_cols = [col for col in housing.columns if col != "MEDV" and not col.startswith("Unnamed")]
x_test = x_test[train_cols]

# Step 4: remove NaN from target
valid_idx = y_test.notna()
x_test = x_test[valid_idx]
y_test = y_test[valid_idx]

# Step 5: transform + predict
x_test_prepared = my_pipeline.transform(x_test)
final_predictions = model.predict(x_test_prepared)

# Step 6: remove NaN from predictions (safety)
valid_idx2 = ~np.isnan(final_predictions)
y_test = y_test[valid_idx2]
final_predictions = final_predictions[valid_idx2]

# Step 7: RMSE
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

print("Final RMSE:", final_rmse)

```
